```python
import pandas as pd
import numpy as np
from konlpy.tag import Okt
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
import pyLDAvis.lda_model
from gensim import corpora
from gensim.models import LdaModel, CoherenceModel
import matplotlib.pyplot as plt

# 1. 텍스트 전처리
def preprocess_text(text):
    okt = Okt()
    tokens = okt.nouns(text)  # 명사만 추출
    return tokens

# content 변수를 사용하여 전처리
df_filter['preprocessed'] = content.apply(preprocess_text)

# 2. gensim 형식으로 데이터 변환
dictionary = corpora.Dictionary(df_filter['preprocessed'])
corpus = [dictionary.doc2bow(text) for text in df_filter['preprocessed']]

# 3. 단어 벡터화 (sklearn)
vectorizer = CountVectorizer(max_df=0.95, min_df=2, stop_words='english')
doc_term_matrix = vectorizer.fit_transform(df_filter['preprocessed'].apply(' '.join))
```

```python
# 4. 토픽 수에 따른 coherence와 perplexity 계산
def compute_coherence_values(dictionary, corpus, texts, limit, start=2, step=2):
    coherence_values = []
    perplexity_values = []
    model_list = []
    for num_topics in range(start, limit, step):
        # Gensim LDA model for coherence
        gensim_model = LdaModel(corpus=corpus, num_topics=num_topics, id2word=dictionary)
        coherencemodel = CoherenceModel(model=gensim_model, texts=texts, dictionary=dictionary, coherence='c_v')
        coherence_values.append(coherencemodel.get_coherence())

        # Sklearn LDA model for perplexity
        sklearn_model = LatentDirichletAllocation(n_components=num_topics, random_state=42)
        sklearn_model.fit(doc_term_matrix)
        perplexity_values.append(sklearn_model.perplexity(doc_term_matrix))
        model_list.append(sklearn_model)

    return model_list, coherence_values, perplexity_values
```

```python
# 5. coherence와 perplexity 계산
model_list, coherence_values, perplexity_values = compute_coherence_values(dictionary=dictionary, corpus=corpus, texts=df_filter['preprocessed'], start=2, limit=20, step=2)

# 6. 결과를 데이터프레임으로 만들기
results_df = pd.DataFrame({
    'Num_Topics': range(2, 20, 2),
    'Coherence_Score': coherence_values,
    'Perplexity_Score': perplexity_values
})

# 결과 표 출력
print(results_df.to_string(index=False))
```

```python
# 7. 최적의 토픽 수 선정 (coherence 기준)
optimal_model = model_list[coherence_values.index(max(coherence_values))]
print(f"\nOptimal number of topics: {optimal_model.n_components}")

# 8. 토픽별 주요 단어 출력
def print_topics(model, vectorizer, top_n=10):
    topics_df = pd.DataFrame()
    for idx, topic in enumerate(model.components_):
        top_words = [(vectorizer.get_feature_names_out()[i], topic[i])
                     for i in topic.argsort()[:-top_n - 1:-1]]
        topics_df[f'Topic {idx+1}'] = [word for word, _ in top_words]
    return topics_df

topics_df = print_topics(optimal_model, vectorizer)
print("\nTop words for each topic:")
print(topics_df.to_string(index=False))
```

```python
# 9. Coherence와 Perplexity 그래프 그리기
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(results_df['Num_Topics'], results_df['Coherence_Score'])
plt.xlabel("Number of Topics")
plt.ylabel("Coherence Score")
plt.title("Coherence Scores by Number of Topics")

plt.subplot(1, 2, 2)
plt.plot(results_df['Num_Topics'], results_df['Perplexity_Score'])
plt.xlabel("Number of Topics")
plt.ylabel("Perplexity Score")
plt.title("Perplexity Scores by Number of Topics")

plt.tight_layout()
plt.show()
```
