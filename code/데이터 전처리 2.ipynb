# 30개씩 끝어서 보기

```python
import pandas as pd
from IPython.display import display, HTML, clear_output
import ipywidgets as widgets

def view_dataframe_chunks(df, chunk_size=30):
    pd.set_option('display.max_colwidth', None)
    pd.set_option('display.max_rows', None)

    total_rows = len(df)
    current_start = [0]  # 리스트로 감싸서 참조로 전달

    def show_chunk(start):
        clear_output(wait=True)
        end = min(start + chunk_size, total_rows)
        print(f"\n--- 행 {start+1}부터 {end}까지 ---")

        html = df.iloc[start:end].to_html(escape=False)
        html = html.replace('<td>', '<td style="white-space: pre-wrap; word-wrap: break-word;">')
        display(HTML(html))

        if end < total_rows:
            display(next_button)
            display(quit_button)
        else:
            print("모든 데이터를 표시했습니다.")

    def on_next_clicked(b):
        current_start[0] += chunk_size
        show_chunk(current_start[0])

    def on_quit_clicked(b):
        clear_output()
        print("데이터 보기를 종료했습니다.")

    next_button = widgets.Button(description="다음")
    next_button.on_click(on_next_clicked)

    quit_button = widgets.Button(description="종료")
    quit_button.on_click(on_quit_clicked)

    show_chunk(current_start[0])

    pd.reset_option('display.max_colwidth')
    pd.reset_option('display.max_rows')
```

# 증상

```python
df1 = df[df['기존분류']=='증상']
df1.info()
```

```python
# 함수 사용
# 스크롤 내리면 다음 30개 행의 정보가 나오고, 종료 버튼도 있습니다.
view_dataframe_chunks(df1)
```

확인 결과 :
 - 기존분류와 Main_Category가 다른 경우는 따로 빼고 될 듯.
 - 두 컬럼의 값이 똑같다고 해도 답변_body의 내용이 아닌 경우가 있음.

```python
def remove_mismatched_categories(df):
    # 기존분류와 Main_Category가 일치하는 행만 선택
    df_cleaned = df[df['기존분류'] == df['Main_Category']]

    # 제거된 행의 수 계산
    removed_rows = len(df) - len(df_cleaned)

    print(f"총 {removed_rows}개의 행이 제거되었습니다.")
    print(f"원본 데이터프레임 크기: {len(df)}")
    print(f"정제된 데이터프레임 크기: {len(df_cleaned)}")

    return df_cleaned

# 함수 사용 예시
df1_cleaned = remove_mismatched_categories(df1)
```

```python
# 결과 확인
view_dataframe_chunks(df1_cleaned)
```

내용이 이상한 행 순서 :

```python
# 제거된 행 확인
removed_rows = df1[df1['기존분류'] != df1['Main_Category']]
print("\n제거된 행:")
print(f"총 {len(removed_rows)}개의 행이 제거되었습니다.")

# view_dataframe_chunks 함수를 사용하여 제거된 행 확인
view_dataframe_chunks(removed_rows)
```
