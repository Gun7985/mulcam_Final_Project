```python
# 필요한 라이브러리 임포트
import pandas as pd  # 데이터 조작 및 분석을 위한 pandas 라이브러리
import numpy as np  # 수치 계산을 위한 numpy 라이브러리
from sentence_transformers import SentenceTransformer  # 문장 임베딩을 위한 SentenceTransformer
import chromadb  # ChromaDB를 위한 라이브러리 (데이터베이스 관련)
from chromadb.utils import embedding_functions  # ChromaDB에서 임베딩 관련 함수 임포트
import torch  # PyTorch 라이브러리, 딥러닝 프레임워크
from transformers import AutoTokenizer, AutoModelForCausalLM  # Transformers 라이브러리에서 자동 토크나이저 및 모델 로드
from kobert_tokenizer import KoBERTTokenizer  # KoBERT 모델을 위한 토크나이저
import os  # 운영 체제와 상호작용을 위한 os 라이브러리
import pickle  # 파이썬 객체를 직렬화하고 역직렬화하기 위한 pickle 라이브러리
from tqdm import tqdm  # 진행 상황 표시를 위한 tqdm 라이브러리
```

```python
# 모델 및 토크나이저 로드 함수 정의
def load_or_create_models(model_dir='./models'):
    # 모델을 저장할 디렉토리가 없으면 생성
    os.makedirs(model_dir, exist_ok=True)

    # 각 모델과 토크나이저의 저장 경로 정의
    sentence_model_path = os.path.join(model_dir, 'sentence_model.pkl')
    kobert_model_path = os.path.join(model_dir, 'kobert_model.pkl')
    tokenizer_path = os.path.join(model_dir, 'tokenizer.pkl')

    # 문장 모델 로드 또는 생성
    if os.path.exists(sentence_model_path):
        # 이미 저장된 문장 모델이 있으면 로드
        with open(sentence_model_path, 'rb') as f:
            sentence_model = pickle.load(f)
    else:
        # 문장 모델이 없으면 새로 생성하고 저장
        sentence_model = SentenceTransformer("jhgan/ko-sroberta-multitask")
        with open(sentence_model_path, 'wb') as f:
            pickle.dump(sentence_model, f)

    # KoBERT 모델과 토크나이저 로드 또는 생성
    if os.path.exists(kobert_model_path) and os.path.exists(tokenizer_path):
        # 이미 저장된 모델과 토크나이저가 있으면 로드
        with open(kobert_model_path, 'rb') as f:
            kobert_model = pickle.load(f)
        with open(tokenizer_path, 'rb') as f:
            tokenizer = pickle.load(f)
    else:
        # 모델과 토크나이저가 없으면 새로 생성하고 저장
        tokenizer = KoBERTTokenizer.from_pretrained('skt/kobert-base-v1')  # KoBERT 토크나이저 로드
        kobert_model = AutoModelForCausalLM.from_pretrained('skt/kobert-base-v1')  # KoBERT 모델 로드
        with open(kobert_model_path, 'wb') as f:
            pickle.dump(kobert_model, f)  # 모델 저장
        with open(tokenizer_path, 'wb') as f:
            pickle.dump(tokenizer, f)  # 토크나이저 저장

    # 로드된 모델과 토크나이저 반환
    return sentence_model, kobert_model, tokenizer
```

```python
# 모델 로드
sentence_model, kobert_model, tokenizer = load_or_create_models()

# Chroma 클라이언트 설정 및 컬렉션 생성/가져오기
chroma_client = chromadb.PersistentClient(path="./chroma_db")
embedding_function = embedding_functions.SentenceTransformerEmbeddingFunction(model_name="jhgan/ko-sroberta-multitask")
collection = chroma_client.get_or_create_collection(name="korean_qa_collection", embedding_function=embedding_function)
```

```python
# 데이터베이스에 데이터 추가 (필요한 경우)
def add_data_to_collection(df, batch_size=1000):
    total_added = 0  # 추가된 데이터의 총 개수를 저장하는 변수 초기화

    # 데이터프레임을 batch_size 단위로 나누어 반복
    for i in tqdm(range(0, len(df), batch_size), desc="데이터 추가 중"):
        # 현재 배치에 해당하는 데이터 선택
        batch = df.iloc[i:i+batch_size]

        # 선택한 배치를 데이터베이스의 컬렉션에 추가
        collection.add(
            documents=batch['답변'].tolist(),  # '답변' 열의 내용을 리스트로 변환하여 문서로 추가
            metadatas=batch[['질병_카테고리', '질병', '부서', '의도']].to_dict('records'),  # 메타데이터를 딕셔너리 형태로 변환
            ids=[f"doc_{j}" for j in range(i, min(i+batch_size, len(df)))]  # 각 문서에 대한 고유 ID 생성
        )

        total_added += len(batch)  # 추가된 배치의 크기를 총 추가된 개수에 더함

    # 추가된 데이터의 총 개수 출력
    print(f"총 {total_added} 개의 데이터가 추가되었습니다.")
```

```python
# 유사 답변 검색 함수
def query_similar_answers(query, n=1):
    try:
        # 주어진 쿼리를 사용하여 데이터베이스에서 유사한 답변 검색
        results = collection.query(query_texts=[query], n_results=n)
        return results  # 검색 결과 반환
    except Exception as e:
        # 오류 발생 시 오류 메시지 출력 및 None 반환
        print(f"오류 발생: {e}")
        return None

# 가장 유사한 답변 가져오기
def get_best_answer(query):
    # 유사 답변 검색 함수 호출
    results = query_similar_answers(query)

    # 검색 결과가 존재하고 거리가 유효한 경우
    if results and results['distances'][0]:
        return {
            'document': results['documents'][0][0],  # 가장 유사한 문서 내용 반환
            'metadata': results['metadatas'][0][0],  # 해당 문서의 메타데이터 반환
            'distance': results['distances'][0][0]    # 유사도 거리 반환
        }

    return None  # 유사한 답변이 없는 경우 None 반환
```

```python
# RAG (Retrieval-Augmented Generation) 함수 정의
@torch.no_grad()  # 그래디언트 계산을 비활성화하여 메모리 사용량 줄이기
def generate_rag_response(query, context, max_length=200, temperature=0.7):
    try:
        # 입력 텍스트 포맷 설정
        input_text = f"질문: {query}\n컨텍스트: {context}\n답변:"

        # 입력 텍스트를 토크나이즈하고 텐서로 변환
        input_ids = tokenizer.encode(input_text, return_tensors='pt', truncation=True, max_length=512)

        # 모델의 최대 길이를 고려하여 생성할 최대 토큰 수 설정
        max_new_tokens = min(max_length, 512 - input_ids.shape[1])

        # KoBERT 모델을 사용하여 응답 생성
        output = kobert_model.generate(
            input_ids,
            max_new_tokens=max_new_tokens,  # 생성할 최대 새 토큰 수
            num_return_sequences=1,  # 반환할 시퀀스 수
            no_repeat_ngram_size=2,  # n-그램 반복 방지 설정
            top_k=50,  # 상위 k개 샘플에서 샘플링
            top_p=0.95,  # 누적 확률이 p 이하인 샘플에서 샘플링
            temperature=temperature,  # 샘플링의 온도 설정
            do_sample=True  # 온도 기반 샘플링을 위해 do_sample=True
        )

        # 생성된 응답을 디코딩하여 반환
        response = tokenizer.decode(output[0], skip_special_tokens=True)
        return response.split("답변:")[-1].strip()  # '답변:' 이후의 텍스트 반환 및 공백 제거
    except Exception as e:
        # 오류 발생 시 메시지 출력 및 기본 오류 응답 반환
        print(f"응답 생성 중 오류 발생: {e}")
        return "죄송합니다. 응답을 생성하는 중에 문제가 발생했습니다."
```

```python
# 챗봇 함수 정의
def chatbot(query):
    # 주어진 질문에 대해 가장 유사한 답변 검색
    best_answer = get_best_answer(query)

    # 유사한 답변이 존재하는 경우
    if best_answer:
        context = best_answer['document']  # 유사한 답변의 문서 내용 가져오기
        rag_response = generate_rag_response(query, context)  # RAG 모델을 사용하여 응답 생성
        return rag_response, best_answer['metadata']  # 생성된 응답과 메타데이터 반환
    else:
        # 유사한 답변이 없는 경우
        return "죄송합니다. 해당 질문에 대한 적절한 답변을 찾을 수 없습니다.", None  # 오류 메시지와 None 반환
```

```python
# 메인 실행 부분
if __name__ == "__main__":
    # 데이터베이스에 데이터 추가 (필요한 경우)
    if collection.count() == 0:  # 데이터베이스에 데이터가 없는 경우
        add_data_to_collection(sample_b)  # 샘플 데이터를 데이터베이스에 추가
    else:
        # 이미 데이터가 존재하는 경우 데이터 개수 출력
        print(f"이미 {collection.count()}개의 데이터가 데이터베이스에 있습니다.")

    # 챗봇 시작 메시지 출력
    print("챗봇을 시작합니다. '종료'를 입력하면 대화를 종료합니다.")

    # 사용자 입력을 받는 무한 루프
    while True:
        user_input = input("사용자: ")  # 사용자 입력 받기
        if user_input.lower() == '종료':  # 사용자가 '종료'라고 입력하면
            print("챗봇을 종료합니다.")  # 종료 메시지 출력
            break  # 루프 종료

        # 챗봇 응답 생성
        response, metadata = chatbot(user_input)
        print("챗봇:", response)  # 챗봇의 응답 출력

        # 메타데이터가 존재하는 경우 관련 정보 출력
        if metadata:
            print("\n관련 정보:")
            print(f"질병 카테고리: {metadata['질병_카테고리']}")
            print(f"질병: {metadata['질병']}")
            print(f"부서: {metadata['부서']}")
            print(f"의도: {metadata['의도']}")
        print()  # 줄 바꿈을 위한 빈 줄 출력
```
