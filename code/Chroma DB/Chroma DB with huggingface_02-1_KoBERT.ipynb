```python
sample_b = sample_a.sample(n=10000, random_state=42)
sample_b.reset_index(drop=True, inplace=True)
```

```python
col = ['질병_카테고리','질병','부서','의도','답변']
sample_b = sample_b[col]
sample_b.head()
```

```python
# 1. 필요 라이브러리
import pandas as pd
from sentence_transformers import SentenceTransformer
import chromadb
from chromadb.utils import embedding_functions
from tqdm import tqdm
import numpy as np
```

```python
# 2. 임베딩 모델 로드
model = SentenceTransformer("jhgan/ko-sroberta-multitask")

# 3. Chroma 클라이언트 설정
chroma_client = chromadb.PersistentClient(path="./chroma_db")

# 4. 컬렉션 생성
embedding_function = embedding_functions.SentenceTransformerEmbeddingFunction(model_name="jhgan/ko-sroberta-multitask")
collection = chroma_client.get_or_create_collection(name="korean_qa_collection", embedding_function=embedding_function)
```

```python
# 5. 데이터 임베딩 및 Chroma DB에 추가
batch_size = 1000  # 한 번에 처리할 데이터 수

all_documents = sample_b['답변'].tolist()
all_embeddings = model.encode(all_documents, show_progress_bar=True, batch_size=32)

for i in tqdm(range(0, len(sample_b), batch_size)):
    batch = sample_b.iloc[i:i+batch_size]

    ids = batch.index.astype(str).tolist()
    documents = batch['답변'].tolist()
    embeddings = all_embeddings[i:i+batch_size].tolist()

    metadata = batch[['질병_카테고리', '질병', '부서', '의도']].to_dict('records')

    collection.add(
        ids=ids,
        documents=documents,
        embeddings=embeddings,
        metadatas=metadata
    )

# 6. 질의 함수 정의
def query_similar_answers(query, n=5):
    results = collection.query(
        query_texts=[query],
        n_results=n
    )
    return results
```

```python
user_query = "손목 수근관 증후군은 통증 완화와 기능 개선을 위해 어떤 방법을 사용해야할까요?"
similar_answers = query_similar_answers(user_query)
print(similar_answers)
```
