```python
# 필요한 라이브러리 임포트
import pandas as pd  # 데이터 조작 및 분석을 위한 pandas 라이브러리
import yaml  # YAML 파일을 읽고 쓰기 위한 yaml 라이브러리
import logging  # 로깅을 위한 logging 라이브러리
import asyncio  # 비동기 프로그래밍을 위한 asyncio 라이브러리
from sentence_transformers import SentenceTransformer  # 문장 임베딩을 위한 SentenceTransformer 클래스
from llama_index.core import VectorStoreIndex, ServiceContext, Document, StorageContext  # llama_index의 핵심 구성 요소들 임포트
from llama_index.core.llms import CustomLLM  # 사용자 정의 LLM (대규모 언어 모델) 클래스 임포트
from llama_index.embeddings.huggingface import HuggingFaceEmbedding  # Hugging Face 임베딩 클래스 임포트
from transformers import AutoTokenizer, AutoModelForCausalLM  # Transformers 라이브러리에서 자동 토크나이저 및 언어 모델 로드
import torch  # PyTorch 라이브러리, 딥러닝 프레임워크
from typing import Any, List, Optional, Literal  # 타입 힌트를 위한 typing 모듈
from pydantic import Field  # Pydantic의 필드 정의를 위한 Field 클래스 임포트
from llama_index.core.llms import (  # llama_index의 LLM 관련 구성 요소들 임포트
    ChatMessage,  # 챗 메시지 클래스
    CompletionResponse,  # 응답 완료 클래스
    CompletionResponseGen,  # 응답 생성기 클래스
    LLMMetadata,  # LLM 메타데이터 클래스
)
```

```python
# 로깅 설정
logging.basicConfig(filename='chatbot.log', level=logging.INFO,
                    format='%(asctime)s - %(levelname)s - %(message)s')  # 로그 파일과 포맷 설정

# config.yaml 파일 로드 (에러 처리 추가)
try:
    with open('config.yaml', 'r') as file:  # config.yaml 파일을 읽기 모드로 열기
        config = yaml.safe_load(file)  # YAML 파일의 내용을 안전하게 로드하여 config에 저장
except FileNotFoundError:
    # config.yaml 파일이 존재하지 않는 경우
    logging.error("config.yaml 파일을 찾을 수 없습니다.")  # 오류 메시지를 로그에 기록
    # 기본 설정값을 사용하여 config 초기화
    config = {
        'max_input_length': 512,  # 최대 입력 길이 설정
        'max_new_tokens': 200,  # 생성할 최대 새 토큰 수 설정
        'top_k': 50,  # 상위 k개 샘플에서 샘플링 설정
        'top_p': 0.95,  # 누적 확률 p 이하에서 샘플링 설정
        'temperature': 0.3  # 샘플링 온도 설정
    }
except yaml.YAMLError as e:
    # YAML 파일 파싱 중 오류 발생 시
    logging.error(f"config.yaml 파일 parsing 에러: {e}")  # 오류 메시지를 로그에 기록
    raise  # 오류를 다시 발생시켜 프로그램 종료
```

```python
# KoBERTLLM 클래스 정의 (CustomLLM을 상속)
class KoBERTLLM(CustomLLM):
    model_name: Literal["skt/kobert-base-v1"] = "skt/kobert-base-v1"  # 사용할 모델 이름
    tokenizer: Any = None  # 토크나이저 초기화
    model: Any = None  # 모델 초기화

    def __init__(self, **data: Any):
        super().__init__(**data)  # 부모 클래스 초기화
        self._load_model()  # 모델 로드

    def _load_model(self):
        # 토크나이저 또는 모델이 None인 경우
        if self.tokenizer is None or self.model is None:
            try:
                # 사전 훈련된 토크나이저 및 모델 로드
                self.tokenizer = AutoTokenizer.from_pretrained(self.model_name)
                self.model = AutoModelForCausalLM.from_pretrained(self.model_name)
                self.model.eval()  # 평가 모드로 설정
                # CUDA 사용 가능 시 모델을 GPU로 이동
                if torch.cuda.is_available():
                    self.model = self.model.to('cuda')
            except Exception as e:
                # 모델 로딩 중 오류 발생 시 로그 기록
                logging.error(f"모델 로딩 중 에러 발생: {e}")
                raise  # 오류를 다시 발생시켜 프로그램 종료

    async def acomplete(self, prompt: str, **kwargs: Any) -> CompletionResponse:
        # 입력 프롬프트를 토크나이즈하여 텐서로 변환
        input_ids = self.tokenizer.encode(prompt, return_tensors="pt", truncation=True, max_length=config['max_input_length'])
        # CUDA 사용 가능 시 입력 텐서를 GPU로 이동
        if torch.cuda.is_available():
            input_ids = input_ids.to('cuda')

        with torch.no_grad():  # 그래디언트 계산 비활성화
            # 모델의 generate 메서드를 비동기적으로 호출
            output = await asyncio.to_thread(
                self.model.generate,
                input_ids,
                max_length=config['max_new_tokens'],
                num_return_sequences=1,
                no_repeat_ngram_size=2,
                top_k=config['top_k'],
                top_p=config['top_p'],
                temperature=config['temperature']
            )

        # 생성된 텍스트 디코딩
        text = self.tokenizer.decode(output[0], skip_special_tokens=True)
        return CompletionResponse(text=text)  # 응답 반환

    def complete(self, prompt: str, **kwargs: Any) -> CompletionResponse:
        # 동기적으로 acomplete 호출
        return asyncio.run(self.acomplete(prompt, **kwargs))

    def stream_complete(self, prompt: str, **kwargs: Any) -> CompletionResponseGen:
        # 스트리밍 완료는 구현되지 않음
        raise NotImplementedError("Stream completion is not implemented for KoBERTLLM")

    @property
    def metadata(self) -> LLMMetadata:
        # 모델의 메타데이터 반환
        return LLMMetadata(
            model_name=self.model_name,
            context_window=config['max_input_length'],
            num_output=config['max_new_tokens'],
        )
```

```python
# 임베딩 모델 설정
embed_model = HuggingFaceEmbedding(model_name="jhgan/ko-sroberta-multitask")  # Hugging Face에서 제공하는 한국어 다중 작업 모델 로드

# LlamaIndex 서비스 컨텍스트 설정
service_context = ServiceContext.from_defaults(
    llm=KoBERTLLM(),  # 사용자 정의 KoBERTLLM 모델 인스턴스 생성
    embed_model=embed_model  # 앞서 설정한 임베딩 모델을 서비스 컨텍스트에 포함
)
```

```python
def dataframe_to_documents(df):
    documents = []  # 문서 객체를 저장할 리스트 초기화
    for _, row in df.iterrows():  # 데이터프레임의 각 행을 반복
        # 각 행의 정보를 텍스트 형식으로 결합
        text = f"{row['질병_카테고리']}\t{row['질병']}\t{row['부서']}\t{row['의도']}\t{row['답변']}"

        # 메타데이터 딕셔너리 생성
        metadata = {
            "질병_카테고리": row['질병_카테고리'],
            "질병": row['질병'],
            "부서": row['부서'],
            "의도": row['의도']
        }

        # Document 객체 생성
        doc = Document(text=text, metadata=metadata)
        documents.append(doc)  # 생성된 Document 객체를 리스트에 추가
    return documents  # 문서 리스트 반환
```

```python
# 인덱스 생성 함수
def create_index(sample_df):
    # 데이터프레임을 문서 객체 리스트로 변환
    documents = dataframe_to_documents(sample_df)
    # 문서 리스트로부터 벡터 스토어 인덱스 생성
    index = VectorStoreIndex.from_documents(documents, service_context=service_context)
    return index  # 생성된 인덱스 반환

async def aquery_index(index, query_text, filter_dict=None):
    # 인덱스를 쿼리 엔진으로 변환
    query_engine = index.as_query_engine()
    # 필터가 제공된 경우
    if filter_dict:
        # 비동기적으로 쿼리 실행
        response = await asyncio.to_thread(query_engine.query, query_text, filter_dict=filter_dict)
    else:
        # 필터 없이 쿼리 실행
        response = await asyncio.to_thread(query_engine.query, query_text)
    return response.response, response.source_nodes  # 응답과 소스 노드 반환

def sanitize_input(user_input: str) -> str:
    # 기본적인 입력 sanitization 함수
    return user_input.strip().replace('<', '&lt;').replace('>', '&gt;')  # 불필요한 공백 제거 및 HTML 태그 변환
```

```python
async def achatbot(index, query):
    # 사용자 입력을 sanitize하여 안전한 쿼리로 변환
    query = sanitize_input(query)
    filter_dict = {}  # 필터를 저장할 딕셔너리 초기화

    # 쿼리에서 특정 필드에 대한 필터 추출
    for field in ["질병_카테고리", "질병", "부서", "의도"]:
        if f"#{field}:" in query:  # 필드가 쿼리에 포함되어 있는지 확인
            # 필드 값 추출
            value = query.split(f"#{field}:")[1].split()[0]  # 필드 값이 공백으로 구분되므로 첫 번째 값을 가져옴
            filter_dict[field] = value  # 필터 딕셔너리에 추가
            query = query.replace(f"#{field}:{value}", "").strip()  # 쿼리에서 필드 값 제거

    # 인덱스에 대해 비동기적으로 쿼리 실행
    response, source_nodes = await aquery_index(index, query, filter_dict)

    # 소스 노드에서 메타데이터 추출
    metadata = source_nodes[0].node.metadata if source_nodes else None  # 소스 노드가 없으면 None
    logging.info(f"Query: {query}, Response: {response}")  # 쿼리와 응답 로그 기록
    return response, metadata  # 응답과 메타데이터 반환
```

```python
async def run_chatbot(index):
    print("챗봇을 시작합니다. '종료'를 입력하면 대화를 종료합니다.")  # 챗봇 시작 메시지 출력
    print("메타데이터 필터링을 사용하려면 '#필드명:값' 형식을 사용하세요. (예: #부서:내과)")  # 필터링 안내 메시지 출력

    while True:  # 무한 루프 시작
        user_input = input("사용자: ")  # 사용자로부터 입력 받기
        if user_input.lower() == '종료':
            print("챗봇을 종료합니다.")
            break  # 루프 종료

        try:
            # 사용자 입력을 기반으로 챗봇 응답 생성
            response, metadata = await achatbot(index, user_input)
            print("챗봇:", response)  # 챗봇 응답 출력

            if metadata:  # 메타데이터가 존재하는 경우
                print("\n관련 정보:")  # 관련 정보 출력 제목
                for key, value in metadata.items():  # 메타데이터 항목 출력
                    print(f"{key}: {value}")
            print()  # 줄 바꿈을 위한 빈 줄 출력
        except Exception as e:
            # 예외 발생 시 오류 로그 기록 및 사용자에게 알림
            logging.error(f"챗봇 응답 생성 중 에러 발생: {e}")
            print("죄송합니다. 응답을 생성하는 중 오류가 발생했습니다.")
```

```python
import logging
import asyncio

def main():
    try:
        sample_b
    except FileNotFoundError:
        logging.error("샘플 데이터 파일을 찾을 수 없습니다.")
        return
    except pd.errors.EmptyDataError:
        logging.error("샘플 데이터 파일이 비어 있습니다.")
        return

    # 인덱스 생성
    index = create_index(sample_b)  # 데이터프레임을 기반으로 인덱스 생성

    # 비동기 이벤트 루프 얻기
    loop = asyncio.get_event_loop()

    # 진행 상황 표시를 위해 tqdm 사용
    total_steps = 100  # 총 단계 수 (예시로 설정)
    for step in tqdm(range(total_steps), desc="Processing"):
        # 실제 처리 작업을 여기에 추가
        await asyncio.sleep(0.1)  # 비동기 작업 시뮬레이션 (예시)

    # 챗봇 실행
    asyncio.run(run_chatbot(index))  # asyncio.run()으로 비동기 함수 실행

if __name__ == "__main__":
    main()
```
