```python
import os
import logging
from dotenv import load_dotenv
import pandas as pd
import numpy as np
from sentence_transformers import SentenceTransformer
import chromadb
from chromadb.utils import embedding_functions
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM
from kobert_tokenizer import KoBERTTokenizer
import pickle
from tqdm import tqdm
import asyncio
import aiohttp
import yaml
from typing import List, Dict
import re
```

```python
# 환경 변수 로드
load_dotenv()

# 설정 파일 로드
with open("config.yaml", "r") as f:
    config = yaml.safe_load(f)
```

```python
# 로깅 설정
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')
logger = logging.getLogger(__name__)

# 모델을 위한 전역 변수
sentence_model = None
kobert_model = None
tokenizer = None
```

```python
# 모델 및 토크나이저 로딩 함수
def load_or_create_models(model_dir=os.getenv("MODEL_DIR", "./models")):
    global sentence_model, kobert_model, tokenizer

    os.makedirs(model_dir, exist_ok=True)

    sentence_model_path = os.path.join(model_dir, 'sentence_model.pkl')
    kobert_model_path = os.path.join(model_dir, 'kobert_model.pkl')
    tokenizer_path = os.path.join(model_dir, 'tokenizer.pkl')

    try:
        if os.path.exists(sentence_model_path):
            with open(sentence_model_path, 'rb') as f:
                sentence_model = pickle.load(f)
        else:
            sentence_model = SentenceTransformer("jhgan/ko-sroberta-multitask")
            with open(sentence_model_path, 'wb') as f:
                pickle.dump(sentence_model, f)

        if os.path.exists(kobert_model_path) and os.path.exists(tokenizer_path):
            with open(kobert_model_path, 'rb') as f:
                kobert_model = pickle.load(f)
            with open(tokenizer_path, 'rb') as f:
                tokenizer = pickle.load(f)
        else:
            tokenizer = KoBERTTokenizer.from_pretrained('skt/kobert-base-v1')
            kobert_model = AutoModelForCausalLM.from_pretrained('skt/kobert-base-v1')
            with open(kobert_model_path, 'wb') as f:
                pickle.dump(kobert_model, f)
            with open(tokenizer_path, 'wb') as f:
                pickle.dump(tokenizer, f)

        logger.info("모델 로딩 성공")
    except Exception as e:
        logger.error(f"모델 로딩 중 오류 발생: {e}")
        raise
```

```python
# Chroma 클라이언트 설정 및 컬렉션 생성/검색
chroma_client = chromadb.PersistentClient(path=os.getenv("CHROMA_DB_PATH", "./chroma_db"))
embedding_function = embedding_functions.SentenceTransformerEmbeddingFunction(model_name="jhgan/ko-sroberta-multitask")
collection = chroma_client.get_or_create_collection(name="korean_qa_collection", embedding_function=embedding_function)
```

```python
# 비동기 데이터 추가 함수
async def add_data_to_collection(df, batch_size=1000):
    total_added = 0
    for i in tqdm(range(0, len(df), batch_size), desc="데이터 추가 중"):
        batch = df.iloc[i:i+batch_size]
        try:
            await asyncio.to_thread(
                collection.add,
                documents=batch['답변'].tolist(),
                metadatas=batch[['질병_카테고리', '질병', '부서', '의도']].to_dict('records'),
                ids=[f"doc_{j}" for j in range(i, min(i+batch_size, len(df)))]
            )
            total_added += len(batch)
        except Exception as e:
            logger.error(f"데이터 배치 추가 중 오류 발생: {e}")
    logger.info(f"총 {total_added}개의 데이터 포인트가 추가됨")
```

```python
# 유사 답변 검색 함수
def query_similar_answers(query, n=1):
    try:
        results = collection.query(query_texts=[query], n_results=n)
        return results
    except Exception as e:
        logger.error(f"유사 답변 검색 중 오류 발생: {e}")
        return None

# 프롬프트 템플릿
PROMPT_TEMPLATE = """의료 전문가로서 다음 질문에 답변해주세요:

질문: {question}

관련 정보:
{context}

답변 (간결하고 정확하게):
"""
```

```python
# 질문 분해 함수
def decompose_question(question: str) -> List[str]:
    # 간단한 규칙 기반 분해 (실제로는 더 복잡한 로직이 필요할 수 있음)
    return re.split(r'[와과]|그리고', question)

# 다중 문서 검색 및 랭킹
def get_relevant_documents(query: str, n: int = 3) -> List[Dict]:
    results = query_similar_answers(query, n)
    if results and results['documents']:
        return [
            {
                'document': doc,
                'metadata': meta,
                'distance': dist
            }
            for doc, meta, dist in zip(results['documents'][0], results['metadatas'][0], results['distances'][0])
        ]
    return []
```

```python
# 동적 프롬프트 생성
def create_dynamic_prompt(question: str, documents: List[Dict]) -> str:
    # 관련 문서 수를 제한하여 프롬프트 길이 조절
    max_docs = 2
    context = "\n".join([f"- {doc['document'][:100]}..." for doc in documents[:max_docs]])
    return PROMPT_TEMPLATE.format(question=question, context=context)

# 자기 일관성 체크
def self_consistency_check(responses: List[str]) -> str:
    # 간단한 구현: 가장 긴 응답을 선택 (실제로는 더 복잡한 로직이 필요)
    return max(responses, key=len)
```

```python
@torch.no_grad()
def generate_advanced_rag_response(query: str) -> str:
    sub_questions = decompose_question(query)
    all_responses = []

    for sub_q in sub_questions:
        relevant_docs = get_relevant_documents(sub_q)
        prompt = create_dynamic_prompt(sub_q, relevant_docs)

        # 입력 길이 제한
        max_length = config['max_input_length'] - config['max_new_tokens'] - 1  # 여유 공간 확보
        input_ids = tokenizer.encode(prompt, return_tensors='pt', truncation=True, max_length=max_length)

        # 여러 번 생성하여 일관성 체크
        responses = []
        for _ in range(3):  # 3번 생성
            try:
                output = kobert_model.generate(
                    input_ids,
                    max_new_tokens=config['max_new_tokens'],
                    num_return_sequences=1,
                    no_repeat_ngram_size=2,
                    top_k=config['top_k'],
                    top_p=config['top_p'],
                    temperature=config['temperature'],
                    do_sample=True
                )
                response = tokenizer.decode(output[0], skip_special_tokens=True)
                responses.append(response.split("답변:")[-1].strip())
            except RuntimeError as e:
                logger.error(f"응답 생성 중 오류 발생: {e}")
                responses.append("응답을 생성하는 중 오류가 발생했습니다.")

        consistent_response = self_consistency_check(responses)
        all_responses.append(consistent_response)

    # 모든 sub-question의 응답을 결합
    final_response = " ".join(all_responses)
    return final_response
```

```python
# 챗봇 함수
def chatbot(query: str) -> tuple:
    response = generate_advanced_rag_response(query)
    # 메타데이터는 더 이상 단일 문서에 연결되지 않으므로, 여기서는 None을 반환
    return response, None

async def main():
    load_or_create_models()

    # 필요한 경우 데이터베이스에 데이터 추가
    if collection.count() == 0:
        df = sample_b  # sample_b가 이 지점 이전에 정의되거나 로드되었는지 확인
        await add_data_to_collection(df)
    else:
        logger.info(f"데이터베이스에 이미 {collection.count()}개의 데이터 포인트가 있습니다")

    print("챗봇을 시작합니다. '종료'를 입력하여 대화를 종료할 수 있습니다.")
    while True:
        user_input = await asyncio.get_event_loop().run_in_executor(None, input, "사용자: ")
        if user_input.lower() == '종료':
            print("챗봇 세션을 종료합니다.")
            break

        response, _ = chatbot(user_input)
        print("챗봇:", response)
        print()  # 가독성을 위한 빈 줄 출력

# 메인 함수 실행
await main()
```
