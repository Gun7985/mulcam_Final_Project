```python
import os
import logging
from dotenv import load_dotenv
import pandas as pd
import numpy as np
from sentence_transformers import SentenceTransformer
import chromadb
from chromadb.utils import embedding_functions
import torch
from transformers import PreTrainedTokenizerFast, BartForConditionalGeneration
import pickle
from tqdm import tqdm
import asyncio
import yaml
from typing import List, Dict
import re
```

```python
# 환경 변수 로드
load_dotenv()

# 설정 파일 로드
with open("config.yaml", "r") as f:
    config = yaml.safe_load(f)
```

```python
# 로깅 설정
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')
logger = logging.getLogger(__name__)

# 모델을 위한 전역 변수
sentence_model = None
kobart_model = None
tokenizer = None
```

```python
# 모델 및 토크나이저 로딩 함수
def load_or_create_models(model_dir=os.getenv("MODEL_DIR", "./models")):
    global sentence_model, kobart_model, tokenizer

    os.makedirs(model_dir, exist_ok=True)

    sentence_model_path = os.path.join(model_dir, 'sentence_model.pkl')
    kobart_model_path = os.path.join(model_dir, 'kobart_model.pkl')
    tokenizer_path = os.path.join(model_dir, 'tokenizer.pkl')

    try:
        if os.path.exists(sentence_model_path):
            with open(sentence_model_path, 'rb') as f:
                sentence_model = pickle.load(f)
        else:
            sentence_model = SentenceTransformer("jhgan/ko-sroberta-multitask")
            with open(sentence_model_path, 'wb') as f:
                pickle.dump(sentence_model, f)

        if os.path.exists(kobart_model_path) and os.path.exists(tokenizer_path):
            with open(kobart_model_path, 'rb') as f:
                kobart_model = pickle.load(f)
            with open(tokenizer_path, 'rb') as f:
                tokenizer = pickle.load(f)
        else:
            tokenizer = PreTrainedTokenizerFast.from_pretrained('gogamza/kobart-base-v2')
            kobart_model = BartForConditionalGeneration.from_pretrained('gogamza/kobart-base-v2')
            with open(kobart_model_path, 'wb') as f:
                pickle.dump(kobart_model, f)
            with open(tokenizer_path, 'wb') as f:
                pickle.dump(tokenizer, f)

        logger.info("모델 로딩 성공")
    except Exception as e:
        logger.error(f"모델 로딩 중 오류 발생: {e}")
        raise
```

```python
# Chroma 클라이언트 설정 및 컬렉션 생성/검색
try:
    chroma_client = chromadb.PersistentClient(path=os.getenv("CHROMA_DB_PATH", "./chroma_db"))
    embedding_function = embedding_functions.SentenceTransformerEmbeddingFunction(model_name="jhgan/ko-sroberta-multitask")
    collection = chroma_client.get_or_create_collection(name="korean_qa_collection", embedding_function=embedding_function)
except Exception as e:
    logger.error(f"Chroma 클라이언트 설정 중 오류 발생: {e}")
    raise
```

```python
# 비동기 데이터 추가 함수
async def add_data_to_collection(df, batch_size=1000):
    total_added = 0
    for i in tqdm(range(0, len(df), batch_size), desc="데이터 추가 중"):
        batch = df.iloc[i:i+batch_size]
        try:
            await asyncio.to_thread(
                collection.add,
                documents=batch['답변'].tolist(),
                metadatas=batch[['질병_카테고리', '질병', '부서', '의도']].to_dict('records'),
                ids=[f"doc_{j}" for j in range(i, min(i+batch_size, len(df)))]
            )
            total_added += len(batch)
        except Exception as e:
            logger.error(f"데이터 배치 추가 중 오류 발생: {e}")
    logger.info(f"총 {total_added}개의 데이터 포인트가 추가됨")
```

```python
def filter_by_metadata(query: str) -> Dict:
    filters = {}

    # 질병 카테고리 필터링
    if any(keyword in query for keyword in ["관절", "척추", "골", "근육"]):
        filters["질병_카테고리"] = "근골격질환"

    # 질병 필터링
    diseases = [
        '손목 수근관 증후군', '강직성 척추염', '골수염', '건초염', '족저 근막염', '척추측만증',
        'VDT 증후군', '골감소증', '요추관 협착증', '관절염', '염좌', '근감소증', '좌골신경통', '골관절염',
        '구루병', '요추 추간판 탈출증', '무지외반증', '아킬레스 건염', '통풍성 관절염', '거북목 증후군',
        '지체장애', '십자 인대 손상', '외반슬', '내반슬', '골육종', '결절종', '통풍', '류마티스 관절염',
        '선천성 편평발', '결정종', '소아마비', '아킬레스 건 파열', '퇴행성 관절염'
    ]
    for disease in diseases:
        if disease.lower() in query.lower():
            filters["질병"] = disease
            break

    # 부서 필터링
    departments = {
        '신경외과': ['신경', '뇌', '척수'],
        '내과': ['내과'],
        '정형외과': ['정형', '뼈', '관절'],
        '외과': ['수술', '외과'],
        '류마티스내과': ['류마티스', '관절염'],
        '재활의학과': ['재활', '물리치료'],
        '마취통증의학과': ['마취', '통증'],
        '소아청소년과': ['소아', '어린이', '청소년'],
        '신경과': ['신경', '두통', '뇌졸중'],
        '안과': ['눈', '시력'],
        '성형외과': ['성형', '미용'],
        '가정의학과': ['가정', '건강검진'],
        '소아청년과': ['소아', '청년']
    }
    for dept, keywords in departments.items():
        if any(keyword in query for keyword in keywords):
            filters["부서"] = dept
            break

    # 의도 필터링
    intentions = ['약물', '재활', '치료', '식이, 생활', '진단', '증상', '원인', '예방', '정의', '검진', '운동']
    for intention in intentions:
        if intention in query:
            filters["의도"] = intention
            break

    return filters
```

```python
# 유사 답변 검색 함수 (메타데이터 필터링 추가)
def query_similar_answers(query, n=3):
    try:
        filters = filter_by_metadata(query)
        results = collection.query(
            query_texts=[query],
            n_results=n,
            where=filters  # 메타데이터 기반 필터 적용
        )
        return results
    except Exception as e:
        logger.error(f"유사 답변 검색 중 오류 발생: {e}")
        return None
```

```python
# 프롬프트 템플릿
PROMPT_TEMPLATE = """의료 전문가로서 다음 질문에 답변해주세요:

질문: {question}

관련 정보:
{context}

답변 (300자 이내로 간결하고 정확하게):
"""
```

```python
# 다중 문서 검색 및 랭킹 (메타데이터 필터링 반영)
def get_relevant_documents(query: str, n: int = 3) -> List[Dict]:
    results = query_similar_answers(query, n)
    if results and results['documents']:
        return [
            {
                'document': doc,
                'metadata': meta,
                'distance': dist
            }
            for doc, meta, dist in zip(results['documents'][0], results['metadatas'][0], results['distances'][0])
        ]
    return []
```

```python
# 동적 프롬프트 생성 (메타데이터 정보 포함)
def create_dynamic_prompt(question: str, documents: List[Dict]) -> str:
    max_docs = 2
    context = "\n".join([
        f"- 문서: {doc['document'][:100]}...\n  카테고리: {doc['metadata'].get('질병_카테고리', '없음')}, 부서: {doc['metadata'].get('부서', '없음')}"
        for doc in documents[:max_docs]
    ])
    return PROMPT_TEMPLATE.format(question=question, context=context)
```

```python
@torch.no_grad()
def generate_response(prompt: str, max_length: int = 300) -> str:
    try:
        input_ids = tokenizer.encode(prompt, return_tensors='pt', truncation=True, max_length=512)
        output = kobart_model.generate(
            input_ids,
            max_length=max_length,
            num_return_sequences=1,
            no_repeat_ngram_size=2,
            top_k=50,
            top_p=0.95,
            temperature=0.7,
            do_sample=True
        )
        response = tokenizer.decode(output[0], skip_special_tokens=True)
        return response.split("답변:")[-1].strip()[:300]  # 300자로 제한
    except Exception as e:
        logger.error(f"응답 생성 중 오류 발생: {e}")
        return "죄송합니다. 응답을 생성하는 중 오류가 발생했습니다."

# 질문 분해 함수 추가
def decompose_question(question: str) -> List[str]:
    # 간단한 규칙 기반 분해
    return re.split(r'[와과]|그리고', question)

def generate_advanced_rag_response(query: str) -> str:
    try:
        sub_questions = decompose_question(query)
        all_responses = []

        for sub_q in sub_questions:
            relevant_docs = get_relevant_documents(sub_q)
            prompt = create_dynamic_prompt(sub_q, relevant_docs)
            response = generate_response(prompt)
            all_responses.append(response)

        final_response = " ".join(all_responses)
        return final_response[:300]  # 최종 응답을 300자로 제한
    except Exception as e:
        logger.error(f"RAG 응답 생성 중 오류 발생: {e}")
        return "죄송합니다. 응답을 생성하는 중 오류가 발생했습니다. 다시 한 번 질문해 주시겠어요?"
```

```python
# 챗봇 함수 수정
def chatbot(query: str) -> str:
    try:
        response = generate_advanced_rag_response(query)
        return response
    except Exception as e:
        logger.error(f"챗봇 응답 생성 중 오류 발생: {e}")
        return "죄송합니다. 응답을 생성하는 중 문제가 발생했습니다. 잠시 후 다시 시도해 주세요."
```

```python
import asyncio
import nest_asyncio
from IPython.display import clear_output

# nest_asyncio를 적용하여 중첩된 이벤트 루프 허용
nest_asyncio.apply()

async def main():
    try:
        load_or_create_models()

        # 필요한 경우 데이터베이스에 데이터 추가
        if collection.count() == 0:
            df = sample_b  # sample_b를 사용
            await add_data_to_collection(df)
        else:
            logger.info(f"데이터베이스에 이미 {collection.count()}개의 데이터 포인트가 있습니다")

        print("챗봇을 시작합니다. '종료'를 입력하여 대화를 종료할 수 있습니다.")
        while True:
            user_input = input("사용자: ")
            if user_input.lower() == '종료':
                print("챗봇 세션을 종료합니다.")
                break

            response = chatbot(user_input)
            print("챗봇:", response)
            print()  # 가독성을 위한 빈 줄 출력
    except Exception as e:
        logger.error(f"메인 함수 실행 중 오류 발생: {e}")
        print("챗봇 실행 중 오류가 발생했습니다. 프로그램을 재시작해 주세요.")

# 메인 함수 실행 (Jupyter Notebook 환경)
await main()
```
